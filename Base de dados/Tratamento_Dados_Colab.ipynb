Trabalhando as tabelas de culturas Temporárias e Permanentes  

```python
import pandas as pd
```

```python
#lendo os arquivos temporarios
df_temp = pd.read_csv("PAM 1612-Temporárias Municípios.csv", sep=";", header=None)
df_perm = pd.read_csv("PAM 1613-Permanentes Municípios.csv", sep=";", header=None)

```

```python
#renomeando colunas
colunas = ['ID', 'Tipo', 'Ano', 'UF', 'CodMun', 'Produto',
           'Quantidade_Produzida', 'Rendimento', 'Valor_Da_Produção',
           'Area_Colhida', 'Area_Plantada']
```

```python
#juntando as planilhas em um df_merge
df_merge = pd.concat([df_temp, df_perm], ignore_index=True)
```

```python
df_merge
```

```python
df_merge.columns = df_merge.iloc[0] # Definindo os nomes das colunas usando os valores da primeira linha (índice 0)
df_merge = df_merge[1:] # Removendo a primeira linha original, que agora é o cabeçalho
df_merge.rename(columns={df_merge.columns[0]: 'ID'}, inplace=True)
```

```python
#excluindo do df as linhas em que o ID estavam vazias, por conta do merge, o head das planilhas acabou passando junto
df_merge = df_merge[df_merge["ID"].notna()]
```

```python
df_merge.head(5)
```

```python
#descobrindo a quantidade de linhas com informações faltantes
numero_de_linhas_com_nulos = len(df_merge[df_merge.isnull().any(axis=1)].index)
print(numero_de_linhas_com_nulos)
```

```python
#trocando nulos por 0
df_merge = df_merge.fillna(0)
numero_de_linhas_com_nulos = len(df_merge[df_merge.isnull().any(axis=1)].index)
print(numero_de_linhas_com_nulos)
```

```python
df_merge.info()
```

```python
#transformando dados e ordenando por CodMun depois Ano
df_merge["id_municipio"] = df_merge["id_municipio"].astype(int)
df_merge["Ano"] = df_merge["Ano"].astype(int)
df_merge = df_merge.sort_values(by=["id_municipio", "Ano"]).reset_index(drop=True)
#vizualizando o resultado
df_merge.head(30)
```

```python
df_merge.info(())
```

```python
#em analise da tabela ja salva foi identificado a duplicidade na coluna Produto descrita como Café(em grão) Total sempre uma linha antes dessa temos a informação Café (em grão) Canephora
df_merge[df_merge['Produto'].str.contains("Total", case=False, na=False)]
```

```python
#excluindo essa linha duplicada do DF
df_merge = df_merge[~df_merge['Produto'].str.contains("Total", case=False, na=False)]
```

```python
#vizualisando a confirmação dessa exclusão
df_merge[df_merge['Produto'].str.contains("Total", case=False, na=False)]
```

```python
#resetando o ID para começar novamente em 1
df_merge["ID"] = range(1, len(df_merge) + 1)
```

```python
#renomeando colunas para facilitar o entendimento
df_merge = df_merge.rename(columns={'Rendimento médio da produção': 'Rendimento médio por hectare', 'Quantidade produzida': 'Quantidade total da produção', 'Valor da produção': 'Valor monetario da produção'})
```

```python
# convertendo dados em int
df_merge['Quantidade total da produção'] = pd.to_numeric(df_merge['Quantidade total da produção'], errors='coerce').astype('Int64')
df_merge['Rendimento médio por hectare'] = pd.to_numeric(df_merge['Rendimento médio por hectare'], errors='coerce').astype('Int64')
df_merge['Valor monetario da produção'] = pd.to_numeric(df_merge['Valor monetario da produção'], errors='coerce').astype('Int64')
df_merge['Área plantada'] = pd.to_numeric(df_merge['Área plantada'], errors='coerce').astype('Int64')
df_merge['Área colhida'] = pd.to_numeric(df_merge['Área colhida'], errors='coerce').astype('Int64')
```

```python
#transformando o valor moterio em reais
df_merge_temp['Valor monetario da produção'] = df_merge['Valor monetario da produção'].astype(float) * 1000
df_merge_temp.head(5)
```

```python
#tabela extraida do site do ibge https://www.ibge.gov.br/explica/codigos-dos-municipios.php
df_municipios = pd.read_excel('/content/RELATORIO_DTB_BRASIL_2024_MUNICIPIOS.xls')
```

```python
#selecionando somentes as colunas de interesse
df_municipios = df_municipios[['Código Município Completo', 'Nome_Município']]
```

```python
#merge do meu df atual e da nova planilha
df_merge_temp = df_merge_temp.merge(
    df_municipios,
    left_on='id_municipio',
    right_on='Código Município Completo',
    how='left'
)
```

```python
#excluir coluna duplicada
df_merge_temp.drop(columns=['Código Município Completo'], inplace=True)
```

```python
#renomeando coluncas
df_merge_temp.rename(columns={'Nome do Município': 'municipio'}, inplace=True)
```

```python

cols = df_merge_temp.columns.tolist()

# Encontra o índice da coluna 'id_municipio'
idx = cols.index('id_municipio')
cols.remove('Nome_Município')

# Insere a coluna 'Nome_Municipio' na posição seguinte a 'id_municipio'
cols.insert(idx + 1, 'Nome_Município')

# Reordena o DataFrame
df_merge_temp = df_merge_temp[cols]
```
